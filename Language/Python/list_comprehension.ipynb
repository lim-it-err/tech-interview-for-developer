```python
import time
import matplotlib.pyplot as plt 
```

```python
def custom_func(data):
    return data * 2
```

```python
first_ret, second_ret = [], []
for i in [10**n for n in range(8)]: # 5자리까지
    print(f"Experiment on {i}th digits")
    lst_original = [i for i in range(i)]
    print(len(lst_original))
    
    lst = lst_original[:]
    start = time.perf_counter_ns()
    sum_first = sum(map(custom_func, lst))
    time_first = time.perf_counter_ns()-start

    lst = lst_original[:]    
    start = time.perf_counter_ns()
    sum_second = sum(list(map(custom_func, lst)))
    time_second = time.perf_counter_ns()-start
    first_ret.append(time_first)
    second_ret.append(time_second)
```

```python
plt.plot(first_ret)
plt.plot(second_ret)
plt.show()
```

```python
plt.plot([((second_ret[i]-first_ret[i])/2**i) for i in range(len(first_ret))])
plt.plot([((second_ret[i]-first_ret[i])/10**i) for i in range(len(first_ret))])

plt.show()
```
